```python
(survey :=
 pl.read_csv("./data/health_survey.csv")
) .head()
```

```python
(reverse :=
 pl.read_csv("./data/ReverseCodingItems.csv",
             columns = ["Needs Reverse Coding?", "Column Name"])
) .head()
```

```python
(reg := {'Strongly Agree': 5,
              'Somewhat Agree':4,
              'Neither Agree nor Disagree':3,
              'Somewhat Disagree':2,
              'Strongly Disagree':1
             }
)
```

```python
(rev := {'Strongly Agree': 1,
              'Somewhat Agree':2,
              'Neither Agree nor Disagree':3,
              'Somewhat Disagree':4,
              'Strongly Disagree':5
             }
)
```

```python
(q_cols :=
 survey.drop('ID').columns
)
```

```python
(result := survey
    .unpivot(on = q_cols,
             index = 'ID',
             variable_name = 'Question',
             value_name = "Response",
            )
    .with_columns(reg_coding = pl.col('Response')
                                 .replace(reg)
                                 .cast(pl.Int64()),
                  rev_coding = pl.col('Response')
                                .replace(rev)
                                .cast(pl.Int64()),
                 )
    .join(reverse,
          left_on = 'Question',
          right_on = 'Column Name',
          how = 'left'
         )
    .with_columns(coding = pl.when(pl.col('Needs Reverse Coding?') == 'Yes')
                             .then('rev_coding')
                             .otherwise('reg_coding'),
                  q_type = pl.col('Question')
                             .str.split('.')
                             .list.get(0),
                 )
    #.filter(pl.col('Needs Reverse Coding?') == 'Yes')
    #.filter(pl.col('Needs Reverse Coding?') == 'No')
    .pivot(on = 'q_type',
           index = 'ID',
           values = 'coding',
           aggregate_function = 'sum',
          )
) #.head()
#['q_type'].unique()
```

```python
result.write_csv('health_survey_summary.csv')
```
